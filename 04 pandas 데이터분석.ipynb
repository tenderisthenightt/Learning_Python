```python
df3 = pd.DataFrame({
    'A':[1,3,4,3,4],
    'B':[2,3,1,2,3],
    'C':[1,5,2,4,4]
})
df3
```

```python
# 각 열의 최대값과 최소값의 차이를 구하시오
df3.apply(lambda x : x.max() - x.min())  # axis=0 은 디폴트라서 생략됨
# 입력값 : 열 하나
# 출력값 :  열의 최대값 - 열의 최소값
```

```python
# 각 행의 최대값과 최소값의 차이를 구하시오
df3.apply(lambda x : x.max() - x.min(), axis=1)
```

```python
# 각 열의 값들의 건수를 추출하시오
df3.apply(pd.value_counts)
# 열이라는거 자체가 시리즈라서 그냥 쓸 수 있음
# 람다도 함수 밸류카운트도 함수라서 하나만 쓰고 싶음
# 밸류카운트는 판다스가 가진 내장 함수라서 pd. 붙이기
```
