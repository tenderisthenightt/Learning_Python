# 막대그래프/ 바차트 (bar chart)
- bar
- barh
- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.bar
- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.barh


```python
# 막대그래프 만들기
# 1. x,y변수
#2. x 축 눈금, y 축 눈금
#3. 그래프 제목, x축 제목, y축 제목


y = [2, 3, 1]
x = np.arange(len(y))


plt.title('Bar Chart')
xlabel = ['가','나','다']
plt.bar(x,y)
plt.xticks(x, xlabel)
## xricks : x축 눈금
plt.yticks(sorted(y))
plt.xlabel('가나다')
plt.ylabel('빈도수')
plt.show()
```

```python
np.random.seed(0)

people = ['몽룡','춘향','방자','향단']
y_pos = np.arange(len(people))

data = np.random.rand(len(people))*10
error = np.random.rand(len(people))

plt.title('수평막대그래프')
plt.barh(y_pos, data, alpha = 0.4)
# alpha : 투명도
plt.yticks(y_pos, people)
plt.xlabel('x라벨')
plt.show()
```

```python
np.random.seed(0)

people = ['몽룡','춘향','방자','향단']
y_pos = np.arange(len(people))

data = np.random.rand(len(people))*10
error = np.random.rand(len(people))

plt.title('수평막대그래프')
plt.barh(y_pos, data,xerr = error,  alpha = 0.4)
# alpha : 투명도
plt.yticks(y_pos, people)
plt.xlabel('x라벨')
plt.show()
```

```python
# 스템 플롯 : 바차트와 유사하지만 폭이 없는 그래프
# 주로 이산확률함수나 자기상관관계를 묘사할 때 사용된다.
x = np.linspace(0.1, 2 *np.pi, 10)
plt.title('Stem Plot')
plt.stem(x, np.cos(x),'-.')
plt.show()
```

# 파이차트(pie chart)

- 카테고리별 값의 상대적인 비교를 해야할 때 주로 사용한다.
- 파이 차트를 그릴 때 원의 형태를 유지할 수 있도록 plt.asix('equal')명령어를 실행해야 한다.
- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.pie


```python
labels = ['맥주','소주','양주','막걸리']
sizes = [25, 45, 20, 10]
colors = ['yellowgreen','lightskyblue','gold','lightcoral']
explode = (0,0.1, 0, 0) # 빠져나오는것
plt.title('내가 간절히 원하는 것은 무엇??')
plt.pie(sizes, explode, labels = labels, colors = colors, autopct = '%1.1f%%', shadow = True, startangle = 90)
plt.axis('equal')
plt.show()
```

# scatter plot: 산점도(산포도)
- 2차원 데이터 즉, 주개의 연속형(실수) 데이터 집합의 상관관계를 파악할 떄 유용한 시각화 도구
- 점 하나의 위치 (x, y)의 값
-http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.scatter

```python
np.random.seed(0)

x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

plt.title('Scatter Plot')
plt.scatter(x, y)
plt.show()
```

```python
# 버블 차트(bubble chart) : 2차원이 아니라 3차원 또는 4 차원의 산점도를 표현할 경우 점 하나의 크기 혹은 색깔을 이용하여,
# 다른데이터 값을 표시한다.(s - 크기, c - 색깔)
N = 30
np.random.seed(0)

x = np.random.rand(N)
y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi * (15* np.random.rand(N)) ** 2

plt.title('Bubble Chart')
plt.scatter(x,y1 , c = y2, s = y3)
plt.show()
```

# Imshow() : 화상(image)데이터를 표현하는 시각화 도구
- 이미지데이터도 행과 열을 가진 2차원 데이터이다.
- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.imshow

```python
from sklearn.datasets import load_digits

digits = load_digits()
digits
```

```python
x = digits.images[0]
x
```

```python
plt.title('손글씨 데이터셋(mnist digits : 0)')
plt.imshow(x, interpolation = 'nearest', cmap = plt.cm.bone_r)

# 눈금선 없애기
# plt.xticks([])
# plt.yticks([])
plt.grid(False)

# 이미지 안에 숫자가  나오는 부분을 조정
plt.subplots_adjust(left = 0.35, right = 0.65, bottom = 0.35, top = 0.65)

plt.show()

```

```python
plt.title('손글씨 데이터셋(mnist digits : 0)')
plt.imshow(x, interpolation = 'nearest', cmap = plt.cm.Pastel2)

# 눈금선 없애기
# plt.xticks([])
# plt.yticks([])
plt.grid(False)

# 이미지 안에 숫자가  나오는 부분을 조정
plt.subplots_adjust(left = 0.35, right = 0.65, bottom = 0.35, top = 0.65)

plt.show()
```

```python
# interpolation 옵션: 
methods = [ 'none', 'nearest', 'bilinear', 'bicubic', 'spline16', 
           'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric', 'catrom',
          'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos']
fig, axes = plt.subplots(3, 6, figsize =(12, 6),
                        subplot_kw={'xticks' : [], 'yticks':[]})
# subplot : 한 도화지를 3행6열로 자르겠다.여기에 한 칸이 0.0, 0.1 ... 으로 생긴 축을 axes 가 받음
## axes , axis 는 다름

for ax, interp_method in zip(axes.flat, methods) : 
    ax.imshow(x, interpolation = interp_method, cmap = plt.cm.bone_r)
    ax.set_title(interp_method)
    # 텍스트 값을 받을 때는 title이 아니라 set_title
    
plt.show()
```

# Seaborn을 분포를 볼때 사용
- http://seaborn.pydata.org
- Seaborn은 Matplotlib을 기반으로 색생과 테마와 통계용 차트등의 기능을 추가한 시각화 패키지
- 기본적인 시각화 기능은 Matplotlib 패키지에 의존, 통계기능은 Statsmodels 패키지에 의존

```python
# 사이트에서 긁어온 거

# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
iris = sns.load_dataset('iris') # 붓꽃 데이터
titanic = sns.load_dataset('titanic') # 타이타닉 호 탑승객 데잍
flights = sns.load_dataset('flights') # 여객 운송 데이터
```

```python
# 1차원 실수 분포 플롯
# rugplot: 데이터 위치를 x축 위에 작은 선분(rug)으로 나타내어 실제데이터 위치를 보여준다.

x = iris.petal_width.values
sns.kdeplot(x)
sns.rugplot(x)
plt.ylim(0,1.0 )
plt.show()
```

```python
titanic_size = titanic.pivot_table(index = 'class', columns = 'sex', aggfunc = 'size')
titanic_size
```

```python
# 2차원 카테고리 형 데이터 :heatmap (상관분석 등)
sns.heatmap(titanic_size,
           cmap ='crest', annot = True, fmt = 'd')
#  cmap =True: 오른쪽에 color map이 생김
# annot = True 개수를 담을꺼니? - fmt : 포멧
```

```python
# 2차원 카테고리 형 데이터: boxplot

sns.boxplot(x = 'day', y = 'total_bill',data = tips)
plt.show
```

```python
# violinplot :분포를 보는 데이터, 커널의 밀도를 히스토그램을 그려준다
sns.violinplot(x = 'day', y = 'total_bill',data = tips)
plt.show
```

```python
# stripplot: violinplot을 점으로 표현
sns.stripplot(x = 'day', y = 'total_bill',data = tips,jitter = True)
plt.show
```

```python
# swarmplot :점이 옆으로 퍼짐

sns.swarmplot(x = 'day', y = 'total_bill',data = tips)
plt.show
```

```python
# swarmplot :점이 옆으로 퍼짐 2 

sns.swarmplot(x = 'day', y = 'total_bill',hue = 'sex', data = tips)
plt.legend(loc = 9)
# loc = 0, 1, 2, 3..., 'best' 등이 있음
plt.show
```
