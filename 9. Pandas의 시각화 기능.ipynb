- pandas의 시리즈나 데이터프레임에 plot이라는 메서드가 내자오디어 있다.

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# mptplotlib의 한글화
mpl.rc('font',family='Malgun Gothic') ## (맑은 고딕)
```

```python
np.random.seed(0)

df = pd.DataFrame(np.random.randn(100,3),
                 index = pd.date_range('2018-1-1', periods = 100),
                 columns = list('ABC')).cumsum()
df.head(10)
```

```python
df.plot()

plt.title('Pandas의 내장된 plot 메서드 사용 예')
plt.xlabel('날짜')
plt.ylabel('Date')
# '-' 글자가 깨짐
plt.rcParams['axes.unicode_minus'] = False # 축 음수표시 오류 해결
plt.show()

```

- plot 메서드에 kind 인수를 통해 다양한 플롯을 시각화 할 수있다.
- bar, ple, hist, kde, box, scatter...

```python
import seaborn as sns

iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
```

```python
# 막대그래프 
iris.sepal_length[:20].plot(kind='bar')
plt.title('iris의 꽃받침의 길이를 이용한 시각화')
plt.xlabel('data')
plt.ylabel('꽃받침의 길이')
plt.show()

# rot = 0 => x축 숫자 돌리기(rotation?)
```

```python
# 막대그래프 2
iris[:5].plot.bar(rot = 0)
plt.title('Iris 붓꽃 데이터의 막대그래프')
plt.xlabel('Data')
plt.ylabel('각 feature의 값')
plt.ylim(0,7)
#plt.legend(loc = 1) ->  위치지정
plt.legend('가나다라')
plt.show()
```

```python
# 품종별로 각 특성의 평균값을 시각화하시오.

df2 = iris.groupby(iris.species).mean()
df2
```

```python
df2.plot.bar(rot = 0)
plt.title('품종별 feature의 평균')
plt.xlabel('품종')
plt.ylabel('품종별 평균')
```

```python
# 객실등급별 승객수의 비율을 시각화
df3 = titanic.pclass.value_counts()
df3
```

```python
df3.plot.pie(autopct = '%1.2f%%')
plt.axis('equal')
plt.show()
```

```python
iris.plot.hist()
plt.show()
# 판다스는 내가 시각화할 객체를 먼저 담음.
```

```python
iris.plot.line()
plt.show()
```

```python
iris.plot.box()
plt.show()
```

```python
iris.boxplot(by='species')
plt.tight_layout(pad = 5,h_pad =10)
plt.show()
```
