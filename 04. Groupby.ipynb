## Groupby(그룹별 집계)
- groupby 매커니즘 : 분할-적용-결합

```python
ipl_data = {'Team':['Riders','Riders','Devils','Devils','Kings','Kings','Kings','Kings',
                    'Riders','Royals','Royals','Riders'],
           'Rank':[1,2,2,3,3,4,1,1,2,4,1,2],
           'Year':[2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
           'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)
df
```

- Series의 데이터 그룹화 : 특정열인덱싱.groupby(기준열인덱싱)
- DataFrame의 데이터 그룹화 : df.groupby('기준이될컬럼명')

```python
# Team별 points의 합을 추출
df.groupby('Team')['Points'].sum()
# 여러개의 값을 묶을 때 sum()
```

```python
# Team과 Year별로 Points 합계를 추출
df.groupby(['Team','Year'])['Points'].sum()
# 컬럼 하나 이상일때 파이썬은 리스트로 묶음
# R은 컴바인으로 묶었듯이
# 00컬럼 ㅁㅁ컬럼만 points 합계를 보고싶다
```

```python
# 이때 00과 ㅁㅁ이 인덱스가 됨
multi_groupby = df.groupby(['Team','Year'])['Points'].sum()
multi_groupby
```

```python
multi_groupby.index
# 멀티인덱스 객체가 된것을 확인
```

```python
# 인덱스가 되엇으니 따로 보는것도 가능
# 상위인덱스로 추출하는것 (팀이름(상위) - 연도(하위))
multi_groupby['Devils':'Kings']
# 같은 리스트 공간에 있을때 : 쓰기도 함
```

```python
# swaplevel : 인덱스 간의 레벨 변경
multi_groupby.swaplevel().sort_index()
```

```python
multi_groupby
```

```python
multi_groupby.sum(level=0)
# 팀별합계.. 레벨연산이 가능하다
# 근데 프로그램이 비추하네
```

- 집계(aggreate) : 그룹화된 통계 정보를 추출할때 사용하는 함수, agg
- 변환(transform)
- 필터(filter) : 특정 정보를 제거하여 보여주는 필터링기능

```python
grouped = df.groupby('Team')
grouped
```

```python
# get_group() : 키값을 기준으로 분할된 데이터프레임 객체를 반환
# 하나의 객체만 데이터프레임으로 보고싶음
grouped.get_group('Riders')
```

```python
# agg() : 다양한 함수를 한번에 적용하는 함수
grouped.agg(min)
```

```python
grouped.agg(np.mean)
# 넘파이도 제공함
```

```python
grouped.transform(max)
# 그룹이 깨지고 개별 최대값을 추출해준다
# 그룹화된 데이터를 상대로 사용하지만 개별 데이터에 대한 집계를 반환해준다
```

```python
z_value = lambda x : (x - x.mean() / x.std())
# 자기자신 - 평균 / 표준편차
# 영희의 시험점수 (830/10000 과 75/100을 비교할때) 기준을 맞춰야하는데
# 모든 데이터 각각에 적용할때 ""transform""을 쓴다 (그룹 표준화)
# 딱맞는 함수가 없을때 lambda 함수로 사용할 수식을 만들어서 갖다씀
grouped.transform(z_value)
```

```python
# 필터(Filter) : 특정 조건으로 데이터를 검색할때 사용하는 함수
df.groupby('Team').filter(lambda x : len(x) >= 3)
# df.groupby('Team') : 그룹 만들기
# filter() : 조건으로 필터링
```

```python
# 팀별로 묶고 points의 최대값이 850을 초과하는 자료만 추출
df.groupby('Team')['Points'].max()
# 팀별로 묶은 다음에 포인트의 최대값을 보여달라
```

```python
df.groupby('Team').filter(lambda x : x['Points'].max() > 850)
# 이렇게 쓰면 됩니다
# 최대값이 850을 넘는 팀들만 보여주는거
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
tips.tail()
```

```python
# 청구서금액 대비 팁의 비율
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips.head()
```

```python
# 수치 데이터의 기초통계량
tips.describe()
```

```python
# 성별로 그룹 설정
# tips.groupby('sex')
# 이렇게만하면 주소가 뜸
tips.groupby('sex').count()
# 보고싶은 기능 함수를 붙여주기
```

```python
tips.groupby('sex').size()
```

```python
# 성별과 흡연유무로 나누어 데이터의 갯수 분석
tips.groupby(['sex' , 'smoker']).size()
```

```python
# 성별에 따른 평균 팁 비율(백분율)
tips.groupby('sex')['tip_pct'].mean()
```

```python
# 흡연여부에 따른 평균 팁 비율
tips.groupby('smoker')['tip_pct'].mean()
```

```python
tips.pivot_table('tip_pct', 'sex', 'smoker', aggfunc='count', margins=True)
```

```python
tips.pivot_table('tip_pct', 'sex', 'smoker')
# 이것만 쓰면 비율이 나옴
# aggfunc='count' 건수가 나오게끔
# margins=True 여백의 미

# 여성흡연자의 팁 비율이 높다 - 라고 해석할 수 있어야
```

```python
tips.groupby('sex')[['tip_pct']].mean()
```

```python
tips.groupby(['sex','smoker'])[['tip_pct']].describe()
```

```python
# 각 그룹에서 가장 많은 팁과 가장 적은 팁의 차이를 확인하시오
tips.groupby(['sex','smoker'])[['tip_pct']]
```

```python
def max_min(x):
    return x.max() - x.min()

tips.groupby(['sex','smoker'])[['tip_pct']].agg(max_min)
```

```python
# 나온값이 안이뻐서 바꿔보기
tips.groupby(['sex', 'smoker'])[['tip']].max() - tips.groupby(['sex', 'smoker'])[['tip']].min()


def max_min(x):
    return x.max()- x.min()
tips.groupby(['sex', 'smoker'])[['tip']].agg(max_min)



tips.groupby(['sex', 'smoker'])[['tip']].agg(lambda x: x.max()-x.min())
```

```python
tips.groupby(['sex','smoker']).agg({'tip':'mean', 'total_bill':max_min})
# 우리가 만든 함수는 앞뒤에 홑따옴표 안붙임
# 이때는 딕셔너리로 담음
# groupby를 잘하면 원하는 내용만 쏙쏙 뽑을때 좋음
```
