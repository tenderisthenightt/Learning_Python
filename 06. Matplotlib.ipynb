## Matplotlib => pyplot 서브패키지
- 라인 플롯(line plot)
- 스캐터 플롯(scatter plot) : 산점도, 산포도
- 바 차트(bar chart)
- 히스토그램(histogram)
- 상자수염그래프(box plot)
- http://matplotlib.org/gallery.html

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# plot(y축), plot(x축, y축)
# 가장 기본적인 라인그래프를 그리는 함수
plt.title('한글쓰면 깨져요')
plt.plot([1, 4, 9, 16])
plt.show()
```

```python
plt.title('x ticks')
# plot(x축, y축)
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show()
```

```python
plt.title('한글 제목')
plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.xlabel('x축 제목')
plt.ylabel('y축 제목')
plt.show()
```

```python
plt.title('한글 제목')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'rs--') 
# 선색(color),볼릿기호(marker),선스타일(line style)
plt.xlabel('x축 제목')
plt.ylabel('y축 제목')
plt.show()
```

```python
plt.plot([10,20,30,40], [1,4,9,16], c='b',
        lw=5, ls='--', marker='o', ms=15, mec='g', mew=5, mfc='r')
# 축범위
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 256)
y = np.cos(x)
plt.plot(x, y)
plt.yticks([-1, 0, 1], ['low','zero','high'])
# 눈금자리에 글자 대체
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.grid(True)
# 보기편하게 눈금선 추가
plt.show()
# ticks 설정한거 확인
```

```python
# 여러개의 선 그래프
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
# (x축, y축, 'x축색깔red&--', 'y축색깔blue', ^ :위쪽 삼각형)
plt.show()
```

```python
#여러개의 선 그래프
t = np.arange(0., 5., 0.2)

plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 256)
c, s = np.cos(x), np.sin(x)
# 패킹  언패킹
plt.plot(x, c, ls='--', label='cosine')
# 스타일 '--' : 실선
plt.plot(x, s, ls=':', label='sine')
# 스타일 ':' : 점선
plt.legend(loc='best')
# 범주 얘가 보기에 최적의 위치에 그려줌
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 256)
c, s = np.cos(x), np.sin(x)
plt.plot(x, c, ls='--', label='cosine')
plt.plot(x, s, ls=':', label='sine')
plt.legend(loc=0)
plt.show()
```
